# Using tensorboard

텐서보드는 좋은 시각화 도구이다 코드에서 무슨일이 일어나는지 확인할 수 있는

예를 들어서 이 예제에서 우리는 반복문을 만들것이고 이것을 텐서보드를 통해서 확인해 볼것이다.

```python
import tensorflow as tf
import numpy as np

raw_data = np.random.normal(10, 1, 100)
```

```python
raw_data
```

moving average 는 다음과 같이 정의 하자.

```python
alpha = tf.constant(0.05)
curr_value = tf.placeholder(tf.float32)
prev_avg = tf.Variable(0.)

update_avg = alpha * curr_value + (1 - alpha) * prev_avg
```

우리가 시각화 시킬 내용은 다음과 같다.

```python
avg_hist = tf.summary.scalar("running_average", update_avg)
value_hist = tf.summary.scalar("incoming_values", curr_value)

merged = tf.summary.merge_all()
writer = tf.summary.FileWriter("./logs")
```

이제 moving average를 수행해보자. 

우리는 merged op를 실행시킬 것이다. 이를 통해 어떻게 변화가 일어나는지 볼 수 있다.

```python
init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    for i in range(len(raw_data)):
        summary_str, curr_avg = sess.run([merged, update_avg], feed_dict={curr_value: raw_data[i]})
        sess.run(tf.assign(prev_avg, curr_avg))
        print(raw_data[i], curr_avg)
        writer.add_summary(summary_str, i)
```

이제 텐서보드를 확인해보자. cmd 창에서 다음을 입력하면 된다. 

$ tensorboard --logdir=path/to/logs
